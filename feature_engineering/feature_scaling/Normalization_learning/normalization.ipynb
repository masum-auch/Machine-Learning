# Learning With Normalization.(Machine Learning)

> Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to changes the values of numeric columns in the dataset to use a common scale, without distorting in the range of values or losing information.

**Type of Normalization**
1. MinMaxScaling
2. Mean Normalization
3. Max Absulate
4. Robust Scaling
